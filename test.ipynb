```python
import os

def list_tree(startpath, indent=""):
    """Recursively print directory tree starting from startpath."""
    items = sorted(os.listdir(startpath))
    for idx, item in enumerate(items):
        path = os.path.join(startpath, item)
        connector = "‚îî‚îÄ‚îÄ " if idx == len(items) - 1 else "‚îú‚îÄ‚îÄ "
        print(indent + connector + item)
        if os.path.isdir(path):
            extension = "    " if idx == len(items) - 1 else "‚îÇ   "
            list_tree(path, indent + extension)

# Example usage:
root_dir = r"E:\The Oven Vibe Static\static\images"
print(root_dir)
list_tree(root_dir)
```

```python
import os
from PIL import Image
import pillow_avif  # ensures AVIF support is registered
import pillow_jxl   # ensures JPEG XL support is registered

# Root images folder
root_dir = r"E:\The Oven Vibe Static\static\images"

# Walk through all subfolders
for subdir, _, files in os.walk(root_dir):
    for file in files:
        if file.lower().endswith(".webp"):
            webp_path = os.path.join(subdir, file)
            base_name = os.path.splitext(webp_path)[0]

            try:
                # Open image
                img = Image.open(webp_path).convert("RGB")

                # Save AVIF (compressed, quality ~80)
                avif_path = base_name + ".avif"
                img.save(avif_path, "AVIF", quality=80)

                # Save JPEG XL (compressed, quality ~90)
                jxl_path = base_name + ".jxl"
                img.save(jxl_path, "JXL", quality=90)

                print(f"‚úÖ {file} -> {os.path.basename(avif_path)}, {os.path.basename(jxl_path)}")

            except Exception as e:
                print(f"‚ùå Failed {file}: {e}")

print("\nüéØ Conversion complete.")
```
